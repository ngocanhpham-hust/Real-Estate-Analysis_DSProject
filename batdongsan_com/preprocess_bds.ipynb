```python
# Batdongsan.com.vn — Data Preprocessing

Input: "data/raw/batdongsan_com_vn.csv"
Output: "data/preprocessed/batdongsan_com_vn_clean.csv"
```

```python
import re
import os
import numpy as np
import pandas as pd
```

```python
RAW_PATH = "data/raw/batdongsan_com_vn.csv"
OUT_PATH = "data/preprocessed/batdongsan_com_vn_clean.csv"

df = pd.read_csv(RAW_PATH, encoding="utf-8")
print(df.shape)
df.head()
```

```python
df.info()
df.isna().mean().sort_values(ascending=False).head(15)
```

```python
# drop cols
cols_to_drop = [
    "balcony_direction",
    "front_road_width",
    "title",
    "verified",
    "date_of_posting",
    "url",
]
df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])
print(df.shape)
df.head(3)
```

```python
# rename col


rename_map = {
    "location_details": "address",
    "front_width": "front_road_width"
}
df = df.rename(columns={k: v for k, v in rename_map.items() if k in df.columns})
df.columns
```

```python
# split location

def split_location(loc: str):
    if pd.isna(loc):
        return pd.Series([np.nan, np.nan])
    parts = [x.strip() for x in str(loc).split(",")]
    if len(parts) >= 2:
        district = parts[0]
        city_province = parts[1]
        return pd.Series([district, city_province])
    return pd.Series([np.nan, np.nan])

df[["district", "city_province"]] = df["location"].apply(split_location)
df = df.drop(columns=["location"])
df[["district", "city_province"]].head(10)
```

```python
def to_float_vn(x):
    """
    Convert Vietnamese number string with comma decimal to float.
    Examples: '7,7' -> 7.7 ; '4,35' -> 4.35
    """
    if pd.isna(x):
        return np.nan
    s = str(x).strip().lower()
    s = s.replace(",", ".")
    try:
        return float(s)
    except:
        return np.nan

def extract_int(x):
    """
    Extract first integer from a string.
    Examples: '2 phòng' -> 2 ; '6 tầng' -> 6
    """
    if pd.isna(x):
        return np.nan
    m = re.search(r"\d+", str(x))
    return float(m.group()) if m else np.nan  # float for NaN compatibility
```

```python
def clean_area_m2(x):
    if pd.isna(x):
        return np.nan
    s = str(x).lower().strip()
    s = s.replace("m²", "").replace("m2", "").replace(" ", "")
    s = s.replace(",", ".")
    s = re.sub(r"[^0-9.]", "", s)
    try:
        return float(s)
    except:
        return np.nan

df["area"] = df["area"].apply(clean_area_m2)
df["area"].describe()
```

```python
for col in ["n_bedrooms", "n_bathrooms", "n_floors"]:
    if col in df.columns:
        df[col] = df[col].apply(extract_int)

df[["n_bedrooms", "n_bathrooms", "n_floors"]].head(10)
```

```python
def clean_width_m(x):
    if pd.isna(x):
        return np.nan
    s = str(x).lower().strip()
    s = s.replace("m", "").replace(" ", "")
    s = s.replace(",", ".")
    s = re.sub(r"[^0-9.]", "", s)
    try:
        return float(s)
    except:
        return np.nan

if "front_road_width" in df.columns:
    df["front_road_width"] = df["front_road_width"].apply(clean_width_m)

df["front_road_width"].describe()
```

```python

# --- normalize categorical text (with villa fix) ---
def normalize_cat(x):
    if pd.isna(x):
        return np.nan

    s = str(x).strip().lower()

    # remove quotes if any
    s = s.strip('"').strip("'")

    # normalize separators/spaces
    s = s.replace("/", "_").replace("-", "_")
    s = re.sub(r"\s+", "_", s)
    s = re.sub(r"_+", "_", s)

    # remove trailing punctuation
    s = s.strip("., ")

    # FIX: map any villa/townhouse variants -> "biệt_thự"
    if ("biệt_thự" in s) or ("liền_kề" in s) or ("lien_ke" in s) or ("villa" in s) or ("townhouse" in s):
        return "biệt_thự"

    return s

cat_cols = ["property_type", "legal", "interior", "facing_direction", "district", "city_province"]
for c in cat_cols:
    if c in df.columns:
        df[c] = df[c].apply(normalize_cat)

df[cat_cols].head(5)
```

```python
# -------------------------------
# Clean & standardize price (billion VND)
# Handles:
# - "x tỷ"
# - "y triệu/m²"
# - invalid cases -> NaN
# -------------------------------

def clean_price_billion(row):
    price_raw = row["price"]
    area = row["area"]

    if pd.isna(price_raw):
        return np.nan

    s = str(price_raw).lower().strip()

    # remove spaces
    s = s.replace(" ", "")

    # -------- Case 1: x tỷ --------
    if "tỷ" in s or "ty" in s:
        s = s.replace("tỷ", "").replace("ty", "")
        s = s.replace(",", ".")
        s = re.sub(r"[^0-9.]", "", s)
        try:
            return float(s)
        except:
            return np.nan

    # -------- Case 2: y triệu / m2 --------
    if "triệu/m" in s or "tr/m" in s or "tr/m2" in s or "triệu/m2" in s:
        if pd.isna(area):
            return np.nan

        s = s.replace("triệu/m²", "").replace("triệu/m2", "")
        s = s.replace("tr/m²", "").replace("tr/m2", "")
        s = s.replace(",", ".")
        s = re.sub(r"[^0-9.]", "", s)

        try:
            price_per_m2_million = float(s)
            total_price_billion = (price_per_m2_million * area) / 1000
            return round(total_price_billion, 3)
        except:
            return np.nan

    # -------- Other cases --------
    return np.nan


df["price"] = df.apply(clean_price_billion, axis=1)

df["price"].describe()
```

```python
# Inspect listings priced per m²
mask_price_per_m2 = df["price"].isna() & df["area"].notna()

df.loc[mask_price_per_m2, ["price", "area"]].head(10)
```

```python
def normalize_interior(x):
    if pd.isna(x):
        return "unknown"

    s = str(x).lower().strip("., ")

    if ("đầy" in s) or ("full" in s) or ("hoàn_thiện" in s):
        return "full"
    if "cơ_bản" in s:
        return "basic"
    if "không" in s:
        return "none"
    return "other"

df["interior"] = df["interior"].apply(normalize_interior)

df["interior"].value_counts()
```

```python
def normalize_legal(x):
    if pd.isna(x):
        return "unknown"

    s = str(x).lower().strip("., ")

    if ("sổ" in s) or ("hồng" in s) or ("đỏ" in s):
        return "ownership"
    if ("hđmb" in s) or ("hợp_đồng" in s) or ("mua_bán" in s):
        return "sale_contract"
    return "other"

df["legal"] = df["legal"].apply(normalize_legal)

df["legal"].value_counts()
```

```python
# Fill n_floors median by property_type (safe-ish for houses; apartments often missing)
if "n_floors" in df.columns and "property_type" in df.columns:
    df["n_floors"] = df.groupby("property_type")["n_floors"].transform(
        lambda x: x.fillna(x.median())
    )

# Keep front_width NaN (avoid bias)
df.isna().mean().sort_values(ascending=False).head(15)
```

```python
# numeric
num_cols = ["price", "area", "n_bedrooms", "n_bathrooms", "n_floors", "front_width", "latitude", "longitude"]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# integer-like columns (nullable Int64)
for c in ["n_bedrooms", "n_bathrooms", "n_floors"]:
    if c in df.columns:
        df[c] = df[c].round().astype("Int64")

df.dtypes
```

```python
DO_LABEL_ENCODING = False  # change to True if you want encoded version

if DO_LABEL_ENCODING:
    from sklearn.preprocessing import LabelEncoder

    df_enc = df.copy()
    cat_cols_ml = ["property_type", "legal", "interior", "facing_direction", "district", "city_province"]
    encoders = {}

    for c in cat_cols_ml:
        if c in df_enc.columns:
            le = LabelEncoder()
            df_enc[c] = df_enc[c].astype(str).fillna("nan")
            df_enc[c] = le.fit_transform(df_enc[c])
            encoders[c] = le

    df_enc.head()
```

```python
# -------------------------------
# Remove old processed file (if exists)
# -------------------------------

import os

if os.path.exists(OUT_PATH):
    os.remove(OUT_PATH)
    print(f"Removed old file: {OUT_PATH}")
else:
    print("No existing file to remove.")
```

```python
# -------------------------------
# Save cleaned dataset (overwrite)
# -------------------------------

os.makedirs(os.path.dirname(OUT_PATH), exist_ok=True)

df.to_csv(OUT_PATH, index=False, encoding="utf-8")
print("Saved new clean file:", OUT_PATH)
print("Final shape:", df.shape)
```
